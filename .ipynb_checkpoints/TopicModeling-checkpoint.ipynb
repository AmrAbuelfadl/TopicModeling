## Adding my own stopwords list to the default ones

```python
dataPath = "tweetsB.txt" # Your text path
text = open(dataPath).read()
nlp = spacy.load("en")
my_stop_words = ['RT', 'people', 'retweet', 'tweet', 'hashtag', 'egy', 'egypt', 'Mubarak', 'jan', 'Egypt', 'Egy', 'trip', 'Expedia', 'video', 'ur', 'u'] # Add your own your stop words list
for stopword in my_stop_words:
    lexeme = nlp.vocab[stopword]
    lexeme.is_stop = True
#print(nlp.vocab["protests"].is_stop)  
```

## Preprocessing steps include removing stopwords, urls, numbers, punctuations and creating biagrams

```python
doc = nlp(text)
texts, article = [], []
for w in doc:
	if w.text != '\n' and not w.is_stop and not w.is_punct and not w.like_num and not w.like_url and w.is_ascii and not w.is_left_punct and not w.is_right_punct and w.lang_=='en' and w.is_alpha :
		article.append(w.lemma_)

	if w.text == '\n':
		texts.append(article)
		article= []
bigram = phrases.Phrases(texts)
texts = [bigram[line] for line in texts]
```

## Implementing LDA model

```python
dictionary = Dictionary(texts)
#dictionary = Dictionary.load("dicTweetEgyA.txt") #Loading dictionary
# dictionary.save("dicTweetsEgyVict.txt") #Saving dictionary
corpus = [dictionary.doc2bow(text) for text in texts]
ldamodel = LdaModel(corpus = corpus, num_topics = 7, id2word = dictionary,random_state = 123, passes = 150, alpha = 0.1, eta = 0.001 )
#print(ldamodel.show_topics())
print(ldamodel.print_topics(num_topics = 7, num_words = 9)) # Print the number of topics you want to retrieve from your documents and the number of words in each topic
#ldamodel.save('LDA') # Saving model
```

## Computing perplexity and coherence scores

```python
print('\nPerplexity: ', ldamodel.log_perplexity(corpus)) 
ldatopics = [[word for word, prob in topic] for topicid, topic in ldamodel.show_topics(formatted=False)]
lda_coherence = CoherenceModel(model = ldamodel, texts=texts, dictionary=dictionary , coherence='c_v', window_size=10).get_coherence() #topics=ldatopics
print('\nCoherence Score: ', lda_coherence)
```

## Computing coherence values and using matplotlib to visualize

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=2):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LdaModel(corpus = corpus, num_topics = num_topics, id2word = dictionary, random_state = 123, passes = 150, alpha = 0.1, eta = 0.001) #passes = 150
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus, texts=texts, start=2, limit=10 , step=2)
limit=10; start=2; step=2;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```
